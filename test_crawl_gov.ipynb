```python
import numpy as np
import pandas as pd

import requests
from urllib3.exceptions import InsecureRequestWarning
from urllib3 import disable_warnings
import json
import time 

disable_warnings(InsecureRequestWarning)
```

```python
def get_ds_tinh():
    page = requests.get("https://www.gdt.gov.vn/TTHKApp/jsp/json.jsp?cmd=GET_DS_TINH", verify=False)
    tinh_dict = json.loads(page.text)
    return tinh_dict
```

```python
ds_tinh = get_ds_tinh()
```

```python
ds_tinh = pd.DataFrame(ds_tinh)
ds_tinh
```

```python
ds_tinh['id'] = ds_tinh['id'].astype(int)
```

```python
ds_tinh.to_csv('ds_tinh.csv', index= False)
```

```python
def get_ds_huyen(maTinh):
    url = "https://www.gdt.gov.vn/TTHKApp/jsp/json.jsp?cmd=GET_DS_HUYEN&maTinh={}".format(maTinh)
    page = requests.get(url, verify=False)
    huyen_dict = json.loads(page.text)
    return huyen_dict
```

```python
def get_ds_xa(maCQThue):
    url = "https://www.gdt.gov.vn/TTHKApp/jsp/json.jsp?cmd=GET_DS_XA&maCQThue={}".format(maCQThue)
    page = requests.get(url, verify=False)
    xa_dict = json.loads(page.text)
    return xa_dict
```

```python
summary_dict = {}
ds_huyen_toan_quoc = []
for tinh in ds_tinh['id']:
    ds_huyen = get_ds_huyen(tinh)
    ds_huyen_toan_quoc = ds_huyen_toan_quoc + ds_huyen
    time.sleep(1)
```

```python
ds_huyen_toan_quoc = pd.DataFrame(ds_huyen_toan_quoc)
ds_huyen_toan_quoc
```

```python
ds_huyen_toan_quoc.to_csv('ds_huyen_toan_quoc.csv', index= False)
```

```python
ds_huyen_toan_quoc = pd.read_csv('ds_huyen_toan_quoc.csv')
ds_huyen_toan_quoc
```

```python
ds_xa_toan_quoc = []
for huyen in ds_huyen_toan_quoc['id']:
    ds_xa = get_ds_xa(huyen)
    ds_xa_toan_quoc = ds_xa_toan_quoc + ds_xa
    time.sleep(5)
```

```python
ds_xa_toan_quoc = pd.DataFrame(ds_xa_toan_quoc)
ds_xa_toan_quoc
```

```python
ds_xa_toan_quoc.to_csv('ds_xa_toan_quoc.csv', index= False)
```

```python
ds_xa_toan_quoc = pd.read_csv('ds_xa_toan_quoc.csv')
ds_xa_toan_quoc
```

```python
ds_full = ds_xa_toan_quoc
ds_full.columns = ['maXa', 'tenXa']
```

```python
ds_full['maHuyen'] = ds_full['maXa'].astype(str).str[:-2].astype(int)
ds_full['maTinh'] = ds_full['maXa'].astype(str).str[:-4].astype(int)
ds_full
```

```python
ds_full = pd.merge(ds_full, ds_huyen_toan_quoc, how='left', left_on='maHuyen', right_on= 'id')[['maXa', 'tenXa', 'maHuyen','title', 'maTinh']]
ds_full.rename(columns = {'title':'tenHuyen'}, inplace = True)
ds_full = pd.merge(ds_full, ds_tinh, how='left', left_on='maTinh', right_on= 'id')[['maXa', 'tenXa', 'maHuyen','tenHuyen', 'maTinh', 'title']]
ds_full.rename(columns = {'title':'tenTinh'}, inplace = True)
ds_full
```
